```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (10, 6)
print("Libraries imported successfully!")
```

```python

file_name = 'Sleep_health_and_lifestyle_dataset.csv'
print(f"Target file: {file_name}")
```

```python

df = pd.read_csv(file_name)
print("Dataset loaded successfully!")
```

```python

print("First 5 rows:")
display(df.head())
```

```python

print("Last 5 rows:")
display(df.tail())
```

```python

rows, cols = df.shape
print(f"The dataset contains {rows} rows and {cols} columns.")
```

```python

print("Columns in the dataset:")
for col in df.columns:
    print(f"- {col}")
    
```

```python

print("Dataset Info:")
df.info()
```

```python

print("Numerical Summary Statistics:")
display(df.describe())
```

```python

print("Categorical Summary Statistics:")
display(df.describe(include=['object']))
```

```python

print("Missing values count per column:")
print(df.isnull().sum())
```

```python

if 'Sleep Disorder' in df.columns:
    df['Sleep Disorder'] = df['Sleep Disorder'].fillna('No Disorder')
    df['Sleep Disorder'] = df['Sleep Disorder'].replace('None', 'No Disorder')
    print("Missing values in 'Sleep Disorder' have been filled with 'No Disorder'.")
```

```python

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python

duplicate_count = df.duplicated().sum()
print(f"Total duplicate rows found: {duplicate_count}")
```

```python

if duplicate_count > 0:
    df = df.drop_duplicates()
    print(f"Dropped {duplicate_count} duplicate rows.")
else:
    print("No duplicates to drop.")
    
print(f"New dataset shape: {df.shape}")
```

```python

if 'Person ID' in df.columns:
    df = df.drop('Person ID', axis=1)
    print("'Person ID' column has been removed.")
```

```python

print("Distribution of Sleep Disorders:")
display(df['Sleep Disorder'].value_counts())
```

```python

disorder_counts = df['Sleep Disorder'].value_counts()

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(disorder_counts.index, disorder_counts.values, color=['#348ABD', '#E24A33', '#988ED5'])
ax.set_title('Distribution of Sleep Disorders', fontweight='bold')
ax.set_xlabel('Sleep Disorder')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(df['Age'], bins=10, color='#8EBA42', edgecolor='black')
ax.set_title('Age Distribution of Participants', fontweight='bold')
ax.set_xlabel('Age')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(df['Sleep Duration'], bins=10, color='#FBC15E', edgecolor='black')
ax.set_title('Distribution of Sleep Duration', fontweight='bold')
ax.set_xlabel('Sleep Duration (hours)')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(8, 5))
ax.scatter(df['Age'], df['Sleep Duration'], alpha=0.7, color='#348ABD')
ax.set_title('Age vs. Sleep Duration', fontweight='bold')
ax.set_xlabel('Age')
ax.set_ylabel('Sleep Duration (hours)')
plt.tight_layout()
plt.show()
```

```python

if 'BMI Category' in df.columns:
    bmi_cats = df['BMI Category'].unique()
    bmi_data = [df[df['BMI Category'] == cat]['Quality of Sleep'].dropna() for cat in bmi_cats]
    
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.boxplot(bmi_data, labels=bmi_cats, patch_artist=True, boxprops=dict(facecolor='#E24A33'))
    ax.set_title('Quality of Sleep by BMI Category', fontweight='bold')
    ax.set_xlabel('BMI Category')
    ax.set_ylabel('Quality of Sleep')
    plt.tight_layout()
    plt.show()
```

```python

numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()
print("Correlation matrix calculated successfully.")
display(corr_matrix)
```

```python

fig, ax = plt.subplots(figsize=(10, 8))
cax = ax.matshow(corr_matrix, cmap='coolwarm')
fig.colorbar(cax)

plt.xticks(range(len(corr_matrix.columns)), corr_matrix.columns, rotation=45, ha='left')
plt.yticks(range(len(corr_matrix.columns)), corr_matrix.columns)

# Add annotations
for i in range(len(corr_matrix.columns)):
    for j in range(len(corr_matrix.columns)):
        ax.text(j, i, f"{corr_matrix.iloc[i, j]:.2f}", ha='center', va='center', 
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.7))

plt.title('Feature Correlation Heatmap', pad=20, fontweight='bold')
plt.grid(False)
plt.tight_layout()
plt.show()
```
